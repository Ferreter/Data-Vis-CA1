```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
file_path= Path.cwd()
file_path
```

```python
file_name="season-1920.csv"
data_19_20 = pd.read_csv(file_path / file_name)
file_name="season-2021.csv"
data_20_21 = pd.read_csv(file_path / file_name)
file_name="season-2122.csv"
data_21_22 = pd.read_csv(file_path / file_name)
```

```python
# Empty array
tables = []

tables = pd.concat([data_19_20, data_20_21, data_21_22])
tables=pd.DataFrame(tables)
tables
```

```python
home = pd.DataFrame({
    "team": tables["HomeTeam"],
    "goals": tables["FTHG"],
    "wins": (tables["FTR"] == "H").astype(int),
    "draws": (tables["FTR"] == "D").astype(int),
    "losses": (tables["FTR"] == "A").astype(int),

    "halftime_goals": tables["HTHG"],
    "halftime_wins": (tables["HTR"] == "H").astype(int),
    "halftime_draws": (tables["HTR"] == "D").astype(int),
    "halftime_losses": (tables["HTR"] == "A").astype(int),

    "shots": tables["HS"],
    "shots_on_target": tables["HST"],
    "fouls_commited": tables["HF"],
    "corners": tables["HC"],
    "yellow_cards": tables["HY"],
    "red_cards": tables["HR"]
})
```

```python
away = pd.DataFrame({
    "team": tables["AwayTeam"],
    "goals": tables["FTAG"],
    "wins": (tables["FTR"] == "A").astype(int),
    "draws": (tables["FTR"] == "D").astype(int),
    "losses": (tables["FTR"] == "H").astype(int),

    "halftime_goals": tables["HTAG"],
    "halftime_wins": (tables["HTR"] == "A").astype(int),
    "halftime_draws": (tables["HTR"] == "D").astype(int),
    "halftime_losses": (tables["HTR"] == "H").astype(int),

    "shots": tables["AS"],
    "shots_on_target": tables["AST"],
    "fouls_commited": tables["AF"],
    "corners": tables["AC"],
    "yellow_cards": tables["AY"],
    "red_cards": tables["AR"]
})
```

```python
combined = pd.concat([home, away])
```

```python
combined_data = combined.groupby("team").sum().reset_index()
```
