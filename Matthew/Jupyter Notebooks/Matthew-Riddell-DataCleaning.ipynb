#### Import libraries

```python
import pandas as pd
import numpy as np
```

#### File path and file loading

```python
# Read each CSV
data_2023 = pd.read_csv("C:/Users/Matty/Documents/College Notes & Assignments/Year 5/Data Visualisation and Insight/CAs/CA1/Data-Vis-CA1/Matthew/Datasets/season_2223.csv")
data_2024 = pd.read_csv("C:/Users/Matty/Documents/College Notes & Assignments/Year 5/Data Visualisation and Insight/CAs/CA1/Data-Vis-CA1/Matthew/Datasets/season_2324.csv")
data_2025 = pd.read_csv("C:/Users/Matty/Documents/College Notes & Assignments/Year 5/Data Visualisation and Insight/CAs/CA1/Data-Vis-CA1/Matthew/Datasets/season_2425.csv")
```

#### Season Column

```python
data_2023["Season"] = "2020 / 2021"
data_2024["Season"] = "2021 / 2022"
data_2025["Season"] = "2022 / 2023"
```

#### Merging all three datasets into one

```python
# Combine datasets
combined_data = pd.concat([data_2023, data_2024, data_2025], ignore_index=True)
```

#### Show total missing values across the dataset

```python
total_missing = combined_data.isna().sum().sum()
print("Total missing values in the dataset:", total_missing)
```

#### List containing possible missing values in the columns 

```python
nonsense_values = ["?", "error", "missing", "Missing", "NaN", "nan", "N/A", "n/a", "--", " "]
```

#### Checking the nonsense_values list across the dataset

This loop goes through the dataset while checking the nonsense values list, 
if a value from the list appears in a column, 
it will increase the count by 1, 
at the end of the loop it will then print out a total per column of nonsense values counted across the dataset.

```python
print("Checking for nonsense values:")
for col in combined_data.columns:
    count = combined_data[col].isin(nonsense_values).sum() 
    if count > 0:
        print(col, "has", count, "nonsense values")
```

#### Replace all nonsense values with NaN

```python
combined_data = combined_data.replace(nonsense_values, np.nan)
```

#### Check again for missing values after replacement

```python
print("Number of missing values after cleaning nonsense values:")
print(combined_data.isna().sum())
```

#### List of columns with numerical data

```python
numeric_columns = ['FTHG', 'FTAG', 
                   'HTHG', 'HTAG', 'HS',
                   'AS', 'HST', 'AST', 
                   'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR']
```

#### Fill missing numeric values with the mean of each column / dropping rows with missing values

```python
for col in numeric_columns:
    if col in combined_data.columns:
        combined_data[col] = pd.to_numeric(combined_data[col], errors='coerce')
        ## mean_value = combined_data[col].mean() 
        ## combined_data[col] = combined_data[col].fillna(mean_value)

print("Shape before dropping rows with missing values:", combined_data.shape)
combined_data = combined_data.dropna()
print("Shape after dropping rows with missing values:", combined_data.shape)

print("Missing values after filling numeric columns with mean:")
print(combined_data.isna().sum())
```

#### Checking for outliers

```python
print("Detecting outliers using IQR method:\n")

for col in numeric_columns:
    if col in combined_data.columns:
        Q1 = combined_data[col].quantile(0.25) # https://www.w3schools.com/Python/pandas/ref_df_quantile.asp
        Q3 = combined_data[col].quantile(0.75) # https://www.w3schools.com/Python/pandas/ref_df_quantile.asp
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR
        
        outliers = combined_data[(combined_data[col] < lower_limit) | (combined_data[col] > upper_limit)]
        print(col, "has", outliers.shape[0], "outliers")
```
